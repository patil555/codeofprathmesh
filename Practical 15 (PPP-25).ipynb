Program to create and traverse a Singly Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert_end(self, data):
        n = Node(data)
        if not self.head:
            self.head = n
        else:
            c = self.head
            while c.next:
                c = c.next
            c.next = n

    def traverse(self):
        c = self.head
        while c:
            print(c.data, end=" -> ")
            c = c.next
        print("None")

sll = SinglyLinkedList()
for i in [10, 20, 30, 40]:
    sll.insert_end(i)
sll.traverse()
```

Program to insert a new node at the beginning

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert_beginning(self, data):
        n = Node(data)
        n.next = self.head
        self.head = n

    def traverse(self):
        c = self.head
        while c:
            print(c.data, end=" -> ")
            c = c.next
        print("None")

sll = SinglyLinkedList()
for i in [20, 30, 40]:
    sll.insert_beginning(i)
sll.insert_beginning(10)
sll.traverse()
```

 Program to delete a node

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert_end(self, data):
        n = Node(data)
        if not self.head:
            self.head = n
        else:
            c = self.head
            while c.next:
                c = c.next
            c.next = n

    def delete_node(self, key):
        c = self.head
        if not c:
            return
        if c.data == key:
            self.head = c.next
            return
        prev = None
        while c and c.data != key:
            prev = c
            c = c.next
        if c:
            prev.next = c.next

    def traverse(self):
        c = self.head
        while c:
            print(c.data, end=" -> ")
            c = c.next
        print("None")


sll = SinglyLinkedList()
for i in [10, 20, 30, 40]:
    sll.insert_end(i)

sll.delete_node(20)
sll.traverse()
```
