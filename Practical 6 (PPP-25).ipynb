```python
cart1 = set()

while True:
    print("\n====== SHOPPING CART SYSTEM (SET) ======")
    print("1. Add product IDs to cart")
    print("2. Display all products in cart")
    print("3. Total unique products count")
    print("4. Highest and Lowest product ID")
    print("5. Search product ID")
    print("6. Remove product ID from cart")
    print("7. Display premium products (>= given value)")
    print("8. Union & Intersection with another cart")
    print("9. Display cart in sorted order")
    print("10. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        n = int(input("How many products to add? "))
        for i in range(n):
            pid = int(input(f"Enter product ID {i+1}: "))
            cart1.add(pid)   # duplicates automatically ignored
        print("Products added successfully!")

    elif choice == 2:
        if len(cart1) == 0:
            print("Cart is empty.")
        else:
            print("Cart Products:", cart1)

    elif choice == 3:
        print("Total Unique Products =", len(cart1))

    elif choice == 4:
        if len(cart1) == 0:
            print("Cart is empty.")
        else:
            print("Highest Product ID =", max(cart1))
            print("Lowest Product ID =", min(cart1))

    elif choice == 5:
        pid = int(input("Enter product ID to search: "))
        if pid in cart1:
            print("Product ID found in cart.")
        else:
            print("Product ID not found.")

    elif choice == 6:
        pid = int(input("Enter product ID to remove: "))
        if pid in cart1:
            cart1.remove(pid)
            print("Product removed successfully!")
        else:
            print("Product ID not found in cart.")

    elif choice == 7:
        if len(cart1) == 0:
            print("Cart is empty.")
        else:
            limit = int(input("Enter premium product ID limit: "))
            premium = {p for p in cart1 if p >= limit}
            print("Premium Products:", premium)

    elif choice == 8:
        cart2 = set()
        n = int(input("How many products in second user's cart? "))
        for i in range(n):
            pid = int(input(f"Enter product ID {i+1}: "))
            cart2.add(pid)

        print("Cart1:", cart1)
        print("Cart2:", cart2)
        print("Union (All unique products):", cart1.union(cart2))
        print("Intersection (Common products):", cart1.intersection(cart2))

    elif choice == 9:
        if len(cart1) == 0:
            print("Cart is empty.")
        else:
            print("Sorted Cart Products:", sorted(cart1))

    elif choice == 10:
        print("Exiting... Thank you!")
        break

    else:
        print("Invalid choice! Try again.")
```
