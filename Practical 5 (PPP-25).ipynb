```python
# Library Inventory System using Tuple

# Step 1: Input book data
n = int(input("Enter number of books: "))
books = ()

for i in range(n):
    value = int(input(f"Enter value of book {i+1}: "))
    books += (value,)

# Menu-driven system
while True:
    print("\n------ Library Inventory Menu ------")
    print("1. Display all book records")
    print("2. Total number of books and average value")
    print("3. Highest and lowest book value")
    print("4. Display premium books")
    print("5. Search for a book value")
    print("6. Count occurrences of a book value")
    print("7. Remove duplicate book records")
    print("8. Display tuple in reverse order")
    print("9. Find index of a book value")
    print("10. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        print("Book Records:", books)

    elif choice == 2:
        total = len(books)
        average = sum(books) / total if total > 0 else 0
        print("Total Books:", total)
        print("Average Value:", average)

    elif choice == 3:
        print("Highest Value:", max(books))
        print("Lowest Value:", min(books))

    elif choice == 4:
        limit = int(input("Enter premium value limit: "))
        premium = tuple(b for b in books if b >= limit)
        print("Premium Books:", premium)

    elif choice == 5:
        search = int(input("Enter value to search: "))
        if search in books:
            print("Book value found.")
        else:
            print("Book value not found.")

    elif choice == 6:
        value = int(input("Enter value to count: "))
        print("Occurrences:", books.count(value))

    elif choice == 7:
        unique_books = tuple(set(books))
        books = unique_books
        print("Duplicates removed.")
        print("Updated Book Records:", books)

    elif choice == 8:
        print("Reversed Tuple:", books[::-1])

    elif choice == 9:
        value = int(input("Enter value to find index: "))
        if value in books:
            print("Index:", books.index(value))
        else:
            print("Value not found.")

    elif choice == 10:
        print("Exiting Library Inventory System. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```
