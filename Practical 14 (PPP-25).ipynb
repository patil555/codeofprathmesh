```python
# Stack Operations: Push, Pop, Display

stack = []

def push():
    element = input("Enter element to push: ")
    stack.append(element)
    print("Element pushed successfully!")

def pop():
    if len(stack) == 0:
        print("Stack is empty! Cannot pop.")
    else:
        element = stack.pop()
        print("Popped element is:", element)

def display():
    if len(stack) == 0:
        print("Stack is empty!")
    else:
        print("Stack elements are:")
        for item in reversed(stack):
            print(item)

while True:
    print("\n--- Stack Menu ---")
    print("1. Push")
    print("2. Pop")
    print("3. Display")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        push()
    elif choice == 2:
        pop()
    elif choice == 3:
        display()
    elif choice == 4:
        print("Exiting program...")
        break
    else:
        print("Invalid choice! Please try again.")
```

```python
# Queue Operations: Insert, Update, Delete, Display

queue = []

def insert():
    element = input("Enter element to insert: ")
    queue.append(element)
    print("Element inserted successfully!")

def delete():
    if len(queue) == 0:
        print("Queue is empty! Cannot delete.")
    else:
        element = queue.pop(0)
        print("Deleted element is:", element)

def update():
    if len(queue) == 0:
        print("Queue is empty! Cannot update.")
    else:
        print("Current Queue:", queue)
        pos = int(input("Enter position to update (starting from 1): "))
        
        if pos < 1 or pos > len(queue):
            print("Invalid position!")
        else:
            new_value = input("Enter new value: ")
            queue[pos - 1] = new_value
            print("Element updated successfully!")

def display():
    if len(queue) == 0:
        print("Queue is empty!")
    else:
        print("Queue elements are:")
        for item in queue:
            print(item)

while True:
    print("\n--- Queue Menu ---")
    print("1. Insert")
    print("2. Delete")
    print("3. Update")
    print("4. Display")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        insert()
    elif choice == 2:
        delete()
    elif choice == 3:
        update()
    elif choice == 4:
        display()
    elif choice == 5:
        print("Exiting program...")
        break
    else:
        print("Invalid choice! Please try again.")
```

```python
# Implementing Stack and Queue in the same program

stack = []
queue = []

# -------- STACK FUNCTIONS --------
def push():
    element = input("Enter element to push into stack: ")
    stack.append(element)
    print("Element pushed into stack.")

def pop_stack():
    if len(stack) == 0:
        print("Stack is empty!")
    else:
        print("Popped element:", stack.pop())

def display_stack():
    if len(stack) == 0:
        print("Stack is empty!")
    else:
        print("Stack elements (Top to Bottom):")
        for item in reversed(stack):
            print(item)

# -------- QUEUE FUNCTIONS --------
def insert_queue():
    element = input("Enter element to insert into queue: ")
    queue.append(element)
    print("Element inserted into queue.")

def delete_queue():
    if len(queue) == 0:
        print("Queue is empty!")
    else:
        print("Deleted element:", queue.pop(0))

def display_queue():
    if len(queue) == 0:
        print("Queue is empty!")
    else:
        print("Queue elements (Front to Rear):")
        for item in queue:
            print(item)

# -------- MAIN MENU --------
while True:
    print("\n------ MAIN MENU ------")
    print("1. Stack Operations")
    print("2. Queue Operations")
    print("3. Exit")

    main_choice = int(input("Enter your choice: "))

    if main_choice == 1:
        print("\n--- Stack Menu ---")
        print("1. Push")
        print("2. Pop")
        print("3. Display")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            push()
        elif choice == 2:
            pop_stack()
        elif choice == 3:
            display_stack()
        else:
            print("Invalid choice!")

    elif main_choice == 2:
        print("\n--- Queue Menu ---")
        print("1. Insert")
        print("2. Delete")
        print("3. Display")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            insert_queue()
        elif choice == 2:
            delete_queue()
        elif choice == 3:
            display_queue()
        else:
            print("Invalid choice!")

    elif main_choice == 3:
        print("Exiting program...")
        break

    else:
        print("Invalid choice!")
```
