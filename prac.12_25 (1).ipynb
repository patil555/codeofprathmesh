1.Program for Linear Search

```python
# Linear Search Program

def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i   # return index if found
    return -1           # return -1 if not found

# Main Program
arr = [10, 25, 30, 45, 60, 75]
key = int(input("Enter number to search: "))

result = linear_search(arr, key)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```

2.Program for Binary Search

```python
def binary_search(arr, key):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            left = mid + 1
        else:
            right = mid - 1

    return -1


# Take input in one line (space separated numbers)
arr = list(map(int, input("Enter sorted numbers (space separated): ").split()))
key = int(input("Enter element to search: "))

result = binary_search(arr, key)

if result != -1:
    print(f"Element found at index {result}")
else:
    print("Element not found")
```

3.Program to search for a character in a string (Linear Search logic)

```python
def linear_search(string, key):
    for i in range(len(string)):
        if string[i] == key:
            return i   # Return index if found
    return -1          # Return -1 if not found


# Taking input from user
text = input("Enter a string: ")
char = input("Enter character to search: ")

result = linear_search(text, char)

if result != -1:
    print(f"Character found at index {result}")
else:
    print("Character not found")
```
