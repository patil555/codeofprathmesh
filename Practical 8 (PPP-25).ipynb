```python
def half_adder(a, b):
    sum_bit = a ^ b
    carry = a & b
    return sum_bit, carry

def full_adder(a, b, cin):
    sum_bit = a ^ b ^ cin
    carry = (a & b) | (b & cin) | (a & cin)
    return sum_bit, carry

a = int(input("Enter bit A (0 or 1): "))
b = int(input("Enter bit B (0 or 1): "))
cin = int(input("Enter Carry-in (0 or 1): "))

hs, hc = half_adder(a, b)
print("Half Adder -> Sum:", hs, "Carry:", hc)

fs, fc = full_adder(a, b, cin)
print("Full Adder -> Sum:", fs, "Carry:", fc)
```
